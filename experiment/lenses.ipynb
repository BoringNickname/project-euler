```python
#ol is object to screen
#ls is lens to screen

def xsq(x,a,c,d):
    return (a*(x**2))+(c*x)+d

def xc (x,a,b,c,d):
    return (a*(x**3))+(b*x**2)+c*x+d

def xf (x,a,b,c,d,e):
    return (a*(x**4))+(b*x**3)+c*x**2+d*x+e


def fn(ol, ls):
    ls = ls*0.01
    ol = (ol+0.85)*0.01 #cuz we use centimeters
    ol = ol - ls
    a = (1/ol+1/ls)**-1
    
    coeffs = curve_fit(xsq,ls,ol, p0 = [1,1,1])[0] #using the curve_fit function
    bestfit = xsq(ls,coeffs[0],coeffs[1],coeffs[2])

    coeffsxc = curve_fit(xf, ls, ol, p0 = [1,1,1,1,1])[0] #using the curve_fit function
    bestfitxc = xf(ls,coeffsxc[0],coeffsxc[1],coeffsxc[2],coeffsxc[3], coeffsxc[4] )
    
    #for a lens of focal length f, the theorie curve
    upred = (1/np.mean(a)-1/ls)**-1
    plt.figure(figsize = (14,6))
#     plt.plot(ls, bestfit, '--', label = 'best fit with quadratic')
#     plt.plot(ls, bestfitxc, '--', label = 'best fit with fourth power ')
    plt.plot(ls, upred, label = 'predicted by theory')
    plt.errorbar(ls,ol, 0.002, 0.002, fmt = 'none', color = 'r', label = 'collected data')
    plt.ylabel('distance from object to  the lens')
    plt.xlabel('distance from lens to the screen')
    plt.title('Theory vs collected data for a lens')
    plt.legend()
    plt.show()
    
    print(a, 'list of focal lenghts, in m')
    print(np.mean(a)*100, 'mean of the set, in cm')
    print(a**-1, 'dioptries')
    print(np.mean(a**-1), 'mean of dioptries')
```

```python
ol = np.array([31.4,31.8,34.9,39.2,43.6,48.3,53,57.7])
ls = np.array([15,20,25,30,35,40,45,50])
fn(ol,ls)
```
