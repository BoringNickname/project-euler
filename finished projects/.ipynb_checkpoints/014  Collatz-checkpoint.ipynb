```python
start = time. time()
num = 123
check = 1000000
test = 10000
i = 1
maxsteps = 0
startnum = 0
while i < test:
    steps = 1
    num = i
    while num != 1:
        if num % 2 == 0:
            num = num / 2
            steps += 1
        else:
            num = (3 * num + 1) / 2
            steps += 2
    if steps > maxsteps:
        maxsteps = steps
        startnum = i
    i += 1
print('under', test, ',', maxsteps, 'is the most steps taken,', startnum, 'being the starting number')
end = time. time()
print(end - start,'s')
```

```python
# #wee program to check if the main thing working correctly
# starting_number = 3**7
# num = starting_number
# steps = 0
# ups, downs = 0, 0
# print('starting number', num)




def upsdowns(base, power):
    num  = base**power
    steps = 0
    ups, downs = 0,0
    while num != 1:
            if num % 2 == 0:
                num = num / 2
                steps += 1
                downs += 1
    #             print('O', 'step', steps,num)
            else:
                num = (3 * num + 1)/2 
                steps += 1
                num = num 
                ups += 1
    #             print('X','step', steps, num)
    # print(steps)
    return ups, downs
```

```python
base = 3
streak = 1
for i in range(1,201):
    if upsdowns(base, i)[0] == upsdowns(base, i-1)[0]-1:
        streak += 1
    else:
        print('STREAK LENGTH:', streak)
        streak = 1
    print('step', i, upsdowns(base, i))
```
