```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lamb = 540*10**-7
d = 2*10**-6
gr = 2*10**-5
th = np.linspace(-60*np.pi,60*np.pi,1000)
inot = 0.1
n = 100
gamma = np.pi/lamb*d*np.sin(th)
beta = np.pi/lamb*gr*np.sin(th)
inten = (inot/n) * ((np.sin(n*gamma)**2)/(np.sin(gamma)**2))*((np.sin(beta)**2)/(beta**2))
# print(inten)
```

```python
figure = plt.scatter(gamma, inten)
# plt.xlim([-10,10])
plt.rcParams['figure.figsize']=(30,20)
plt.grid()
plt.show()
```
